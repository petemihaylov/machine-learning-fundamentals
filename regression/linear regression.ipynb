```python
from pydataset import data
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

```python
%matplotlib notebook
```

```python
pima = data("Pima.tr")
print(pima.head())
```

```python
pima.plot(kind="scatter", x="skin", y="bmi")
plt.show()
```

```python
# Test train and split for supervised training
X_train, X_test, y_train, y_test = train_test_split(pima.skin, pima.bmi)
```

```python
plt.scatter(X_train, y_train, label="Training data", color="r", alpha=0.7)
plt.scatter(X_test, y_test, label="Test", color="g", alpha= .7)
plt.legend()
plt.title("Test Train Split")
plt.show()
```

```python
# Create linear model to train
LR = LinearRegression()
LR.fit(X_train.values.reshape(-1,1), y_train.values)
```

```python
# Use the model to predict 
prediction = LR.predict(X_test.values.reshape(-1,1))

# Plot the prediction agains the actual test data
plt.plot(X_test, prediction, label="Linear Regression", color="b")
plt.scatter(X_test, y_test, label="Actual test data", color="g", alpha=0.7)
plt.legend()
plt.show()
```

```python
# Predict BMI from a woman with skin 50
LR.predict(np.array([[50]]))[0]
```

```python
# Score this model 
LR.score(X_test.values.reshape(-1,1), y_test.values)
```
